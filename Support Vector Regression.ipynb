#**Salary Prediction Using Position Level**


> This project aims to predict salary based on given level of position. we had used **Support Vector Machine Regression** in this notebook. 



### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Getting Dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, -1].values
```

```python
# Reshaping y into vector 
y = y.reshape(len(y), 1)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
x = sc_x.fit_transform(x)
y = sc_y.fit_transform(y)
```

### Fitting SVR To Dataset

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(x, y)
```

### Predicting Values With SVR Regressor

```python
y_pred = sc_y.inverse_transform(regressor.predict(sc_x.transform([[6.5]])))
```

### Visualising SVR Model

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.title('Truth or Bluff!! (SVR)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```
