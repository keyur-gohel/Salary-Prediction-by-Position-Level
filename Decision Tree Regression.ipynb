#**Salary Prediction Using Position Level**


> This project aims to predict salary based on given level of position. we had used **Decision Tree Regression** in this notebook. 



### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Getting Dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, -1].values
```

### Fitting Decision Tree Regression In Dataset

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(x, y)
```

### Predicting Value With Decision Tree Regressor

```python
y_pred = regressor.predict([[6.5]])
```

### Visualising Decision Tree Regression Model (Smooth Curve and High Resolution)

```python
x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, regressor.predict(x_grid), color = 'blue')
plt.title('Truth or Bluff!! (Decision Tree)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```
