#**Salary Prediction Using Position Level**


> This project aims to predict salary based on given level of position. we had used **Polynomial Regression** in this notebook. 



### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Getting Dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:2].values
y = dataset.iloc[:,-1].values
```

### Fitting Linear Regression To Data

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
```

### Fitting Polynomial Regression To Data

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg =  PolynomialFeatures(degree = 5)
x_poly = poly_reg.fit_transform(x)
poly_reg.fit(x_poly, y)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly, y)
```

### Visualise Linear Regression Result

```python
plt.scatter(x, y, color =  'red')
plt.plot(x, lin_reg.predict(x), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

### Visualise Polynomial Regression Result

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg2.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

### Visualise Polynomial Regression Result (Smoth curve for bater visualisation)



```python
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape(len(x_grid), 1)
plt.scatter(x, y, color =  'red')
plt.plot(x_grid, lin_reg2.predict(poly_reg.fit_transform(x_grid)), color = 'blue')
#plt.plot(x, lin_reg2.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.title('Truth or Bluff (Polnomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

### Predicting The Salary By linear Regresssion

```python
lin_reg.predict([[6.5]])
```

### Predicting The Salary By Polynomial Regresssion



```python
lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
```
